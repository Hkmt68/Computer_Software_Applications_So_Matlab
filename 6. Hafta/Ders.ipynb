```matlab
% Rastgele noktalar oluturalm
x = rand(100,1);
y = rand(100,1);
z = rand(100,1);

% scatter3 kullanarak dalm grafii izelim
scatter3(x, y, z, 50, 'filled'); % 50 boyutunda dolu daireler kullanarak noktalar izelim
xlabel('X Ekseni');
ylabel('Y Ekseni');
zlabel('Z Ekseni');
title(' Boyutlu Dalm Grafii');
```

```matlab
% Veri oluturalm
sizes = [20, 30, 50];
labels = {'Yeil', 'Turkuaz', 'Sar'};
colors = {'r', 'g', 'b'}; % Renk kodlarn kullanarak renkleri belirleyelim

% pie3 kullanarak  boyutlu pasta grafii izelim
h = pie3(sizes, labels);

% Her dilim iin renkleri ayarlayalm
for i = 1:numel(sizes)
    h((i-1)*2+1).FaceColor = colors{i};
end

title(' Boyutlu Pasta Grafii');
```
