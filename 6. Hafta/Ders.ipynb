```matlab
% Rastgele noktalar oluşturalım
x = rand(100,1);
y = rand(100,1);
z = rand(100,1);

% scatter3 kullanarak dağılım grafiği çizelim
scatter3(x, y, z, 50, 'filled'); % 50 boyutunda dolu daireler kullanarak noktaları çizelim
xlabel('X Ekseni');
ylabel('Y Ekseni');
zlabel('Z Ekseni');
title('Üç Boyutlu Dağılım Grafiği');
```

```matlab
% Veri oluşturalım
sizes = [20, 30, 50];
labels = {'Yeşil', 'Turkuaz', 'Sarı'};
colors = {'r', 'g', 'b'}; % Renk kodlarını kullanarak renkleri belirleyelim

% pie3 kullanarak üç boyutlu pasta grafiği çizelim
h = pie3(sizes, labels);

% Her dilim için renkleri ayarlayalım
for i = 1:numel(sizes)
    h((i-1)*2+1).FaceColor = colors{i};
end

title('Üç Boyutlu Pasta Grafiği');
```
