2. Soru

```matlab
% Fonksiyon tanımı
f = @(x) ( x.^3 + x.^2 - 3*x + 1 )

% x değerleri
x = -6:0.01:6
```

```matlab
% Birinci mertebeden Türev
f_prime = diff(f(x))./0.01 % Merkezi fark yöntemiyle hesaplanır

% İkinci mertebeden Türev
f_double_prime = diff(f_prime)./0.01 % Merkezi fark yöntemiyle hesaplanır
```

```matlab
% Plot komutu
subplot(2,1,1)
plot(x, f(x), 'b', x(1:end-1), f_prime, 'r')
title('Fonksiyon ve Birinci Mertebeden Türevi')
xlabel('x')
ylabel('f(x), f''(x)')
legend('f(x)', 'f''(x)')
grid on

subplot(2,1,2)
plot(x(1:end-2), f_double_prime, 'g')
title('İkinci Mertebeden Türev')
xlabel('x')
ylabel('f''''(x)')
grid on
```

3. Soru

```matlab
% Zaman aralığı tanımlama
t = 0:0.01:2*(1/3) % 2 periyot için zaman aralığı,
% frekansın tersi kadar periyot alınır

% Sinüs sinyali oluşturma
A = 5 % Maksimum genlik
f = 3 % Frekans
y = A * sin(2*pi*f*t)
```

```matlab
% Grafiği Çizme
plot(t,y)
xlabel('Zaman ( s )')
ylabel('Genlik')
title('2 Periyotluk Sinüs Sinyali')
grid on
```

4. Soru

```matlab
figure;

% Grafik 1: u(t) = 5 sin(4t)
subplot(3,1,1);
t = linspace(0, 2*pi, 100);
u1 = 5*sin(4*t);
plot(t, u1, 'b', 'LineWidth', 3);
title('\bf \it u(t) = 5 sin(4t)', 'FontSize', 13, 'FontWeight', 'bold', 'FontAngle', 'italic');
xlabel('\bf Zaman', 'FontSize', 14, 'FontWeight', 'bold');
ylabel('\bf \it u(t)', 'FontSize', 12, 'FontWeight', 'bold');
grid on;

% Grafik 2: u(t) = 3 cos(4t)
subplot(3,1,2);
u2 = 3*cos(4*t);
plot(t, u2, 'r--', 'LineWidth', 2, 'Marker', 'o');
title('\bf \it u(t) = 3 cos(4t)', 'FontSize', 13, 'FontWeight', 'bold', 'FontAngle', 'italic');
xlabel('\bf Zaman', 'FontSize', 14, 'FontWeight', 'bold');
ylabel('\bf\it u(t)', 'FontSize', 12, 'FontWeight', 'bold');
grid on;

% Grafik 3: u1(t) + u2(t)
subplot(3,1,3);
u_sum = u1 + u2;
plot(t, u_sum, 'g:', 'LineWidth', 4, 'Marker', '.');
title('\bf \it u1(t) + u2(t)', 'FontSize', 13, 'FontWeight', 'bold', 'FontAngle', 'italic');
xlabel('\bf Zaman', 'FontSize', 14, 'FontWeight', 'bold');
ylabel('\bf \it u(t)', 'FontSize', 12, 'FontWeight', 'bold');
grid on;

% X ekseni
set(gca, 'XColor', 'k', 'FontSize', 14);

% Y ekseni
set(gca, 'YColor', 'k', 'FontSize', 12, 'FontAngle', 'italic');

% Başlık
sgtitle('\bf\itGrafikler', 'FontSize', 13, 'FontWeight', 'bold', 'FontAngle', 'italic');
```
