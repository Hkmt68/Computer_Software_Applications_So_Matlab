$$ D = \begin{vmatrix} 2 & 4 \\ 1 & 3 \end{vmatrix} $$
$$ D_1 = \begin{vmatrix} 14 & 4 \\ -8 & 3 \end{vmatrix} $$
$$ D_2 = \begin{vmatrix} 2 & 14 \\ 1 & -8 \end{vmatrix} $$

Şimdi determinantsız kısımları hesaplayalım:

$ D = (2*3) - (4*1) = 6 - 4 = 2 $
$ D_1 = (14*3) - (4*-8) = 42 + 32 = 74 $
$ D_2 = (2*-8) - (14*1) = -16 - 14 = -30 $

$ I_1 = \frac{D_1}{D} = \frac{74}{2} = 37 $
$ I_2 = \frac{D_2}{D} = \frac{-30}{2} = -15 $
